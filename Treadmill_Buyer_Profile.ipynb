```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
```

```python
file_path = '/content/drive/My Drive/Datasets/treadmill.csv'
data = pd.read_csv(file_path)
```

# ----------- Data Visualization ------------ #

# Univariate Analysis (Numerical & Categorical)

# 1. Numerical Features Analysis: Univariate
# We will focus on 'Age', 'Education', 'Income', and 'Miles'



```python
plt.figure(figsize=(14, 8))

# Count plot of 'Age'
plt.subplot(2, 2, 1)
sns.countplot(data=data, x='Age', palette='coolwarm')
plt.title('Count Plot of Age')

# Box plot of 'Age'
plt.subplot(2, 2, 2)
sns.boxplot(data=data, x='Age', palette='Set3')
plt.title('Box Plot of Age')

# Count plot of 'Education'
plt.subplot(2, 2, 3)
sns.countplot(data=data, x='Education', palette='coolwarm')
plt.title('Count Plot of Education')

# Box plot of 'Education'
plt.subplot(2, 2, 4)
sns.boxplot(data=data, x='Education', palette='Set3')
plt.title('Box Plot of Education')

plt.tight_layout()
plt.show()

# For 'Income' and 'Miles' (Numerical features)
plt.figure(figsize=(12, 6))

# Count plot of 'Income'
plt.subplot(1, 2, 1)
sns.countplot(data=data, x='Income', palette='Blues')
plt.title('Count Plot of Income')
plt.xticks(rotation=45)

# Count plot of 'Miles'
plt.subplot(1, 2, 2)
sns.countplot(data=data, x='Miles', palette='Greens')
plt.title('Count Plot of Miles')
plt.xticks(rotation=45)
plt.show()
```

# 2. Categorical Features Analysis: Univariate
# Count plot for 'Gender' and 'MaritalStatus'


```python
plt.figure(figsize=(10, 6))

# Count plot of 'Gender'
plt.subplot(1, 2, 1)
sns.countplot(data=data, x='Gender', palette='Set2')
plt.title('Count Plot of Gender')

# Count plot of 'MaritalStatus'
plt.subplot(1, 2, 2)
sns.countplot(data=data, x='MaritalStatus', palette='Set3')
plt.title('Count Plot of Marital Status')

plt.tight_layout()
plt.show()
```

# BIVARIATE ANALYSIS

```python
# Bivariate Analysis: Relationship between 'Income' and 'Age'
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age', y='Income', data=data, palette='coolwarm')
plt.title('Bivariate Analysis of Income vs Age')
plt.show()

# Bivariate Analysis: Relationship between 'Gender' and 'Miles'
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Miles', data=data, palette='Set2')
plt.title('Bivariate Analysis of Miles vs Gender')
plt.show()

# Bivariate Analysis: Relationship between 'Fitness' and 'Income'
plt.figure(figsize=(10, 6))
sns.boxplot(x='Fitness', y='Income', data=data, palette='Set1')
plt.title('Bivariate Analysis of Fitness vs Income')
plt.show()
```

# MULTI VARIATE ANALYSIS

```python
# Multivariate Analysis: Using pairplot for Age, Income, and Miles
sns.pairplot(data[['Age', 'Income', 'Miles']], palette='coolwarm')
plt.title('Pairplot of Age, Income, and Miles')
plt.show()
```

# CORRELATION ANALYSIS

```python
# Correlation Analysis between numerical features
corr = data[['Age', 'Education', 'Income', 'Miles']].corr()

# Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap between Numerical Features')
plt.show()
```

# HEAT MAPS

```python
# Heatmap for Education, MaritalStatus, and Gender
pivot_table = pd.crosstab(index=data['Education'], columns=[data['MaritalStatus'], data['Gender']])
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, annot=True, cmap='viridis', linewidths=0.5)
plt.title('Heatmap of Education vs MaritalStatus and Gender')
plt.show()
```

# PAIR PLOT ANALYSIS

```python
# Pairplot for numerical variables
plt.figure(figsize=(10, 6))
sns.pairplot(data[['Age', 'Income', 'Miles']])
plt.title('Pairplot of Age, Income, and Miles')
plt.show()
```

# Chi-Square Test

```python
# Contingency table for Product and Gender
contingency_table = pd.crosstab(data['Product'], data['Gender'])

# Marginal probabilities for Gender
marginal_gender = contingency_table.sum(axis=1) / len(data)

# Chi-Square Test for Independence
chi2, p, dof, expected = chi2_contingency(contingency_table)
print("\nChi-Square Test for Independence")
print(f"Chi2 Statistic: {chi2}, p-value: {p}")
```
